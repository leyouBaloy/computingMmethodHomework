```python
import matplotlib.pyplot as plt
import numpy as np
plt.figure(figsize=(10,10))
x0=[1.5,1.55,1.6,1.65]
res=[]
cnt=1
x_axis=list(np.arange(1,13))
for x1 in x0:
    res.append(x1)
    x=x1
    for i in range(11):
        x=x**2-1
        res.append(x)
    plt.subplot(2, 2, cnt)
    plt.plot(x_axis,res[(cnt-1)*(i+2):cnt*(i+2)])
    cnt+=1
```
